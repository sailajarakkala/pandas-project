```python
import pandas as pd
df = pd.read_excel("Movies_Dataset.xlsx")
df
```

```python
df.to_excel("MOVIE.xlsx")
df
```

## 1.select all Movies in Action

```python
df = df[df['Genre'] == 'Action']
df
```

## 2.Movies Released after 2000

```python
Movies_after_2000 = df[df['Release_Year']> 2000]
Movies_after_2000
```

## 3.Movies with an IMDb rating greater

```python
IMDb_Rating = df[df['IMDb_Rating']> 8.5]
IMDb_Rating
```

## 4.select only Movie Name and Genre

```python
Movie_Name_Genre = df[["Movie Name","Genre"]]
Movie_Name_Genre
```

## 5.Movies released between 1990 and 2010

```python
Movie_Relesed = df[(df['Release_Year'] >= 1990) & (df['Release_Year'] <= 2010)]
Movie_Relesed
```

## 6.Count the number of movies in each genre.

```python
count_movies = df["Genre"].count()
count_movies
```

## 7.The highest IMDb-rated movie

```python
highest_rated_movie = df.loc[df['IMDb_Rating'].idxmax()]
highest_rated_movie
```

## 8.The avg imdb rating of all movies

```python
avg_rating = df['IMDb_Rating'].mean()
avg_rating 
```

## 9.Find the total box office collection per genre.

```python
total_collection = df.groupby('Genre')['Revenue'].sum().sort_values(ascending=False)
total_collection
```

## 10.Movies with box office earnings greater than 500 million.

```python
high_earning_movies = df[df['Revenue'] > 500_000_000]
high_earning_movies
```

## 11.Rank movies based on IMDb rating

```python
df = df.sort_values(by='IMDb_Rating', ascending=False)
df['Rank'] = range(1, len(df) + 1)
df['Rank']
```

## 12.Sort movies by relese in year decending order

```python
sort_movies = df.sort_values("Release_Year",ascending = False)
sort_movies
```

## 13.Find all duplicate movie names (if any).

```python

duplicate_movies = df[df.duplicated(subset = ["Movie Name"])]
duplicate_movies
```

## 14.Remove duplicate movie names.

```python
remove_deplicates = df.drop(columns = ["Movie Name"])
remove_deplicates 
```

## 15.Replace "sci-fi" genre with "science fiction"

```python
df["Genre"] = df.Genre.replace("sci-fi","science Fiction")
df
```
